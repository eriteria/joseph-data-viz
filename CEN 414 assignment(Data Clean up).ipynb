### Code for cleaning the Data

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
scipy
```

```python
data = pd.read_csv("country_code_web.csv")
country_code = pd.read_csv("country_code_web_main.csv")
```

```python
data.head()
```

```python
country_code.head()
```

```python
ddata = data[['SpatialDimType','SpatialDim', 'TimeDim', 'NumericValue']]
```

```python
ddata.head()
```

```python
ddata = ddata[ddata['TimeDim'] >= 2010]
```

```python
ddata = ddata[ddata['SpatialDimType'] == "COUNTRY"]
```

```python
ddata = ddata[ddata['NumericValue'] > 1]
```

```python
country_code_dict = country_code.set_index('Country or area name')['ISO ALPHA-3 code'].to_dict()
```

```python
country = []
for index, row in ddata.iterrows():
    if row['SpatialDim'] in country_code['ISO ALPHA-3 code'].tolist():
        for key, value in country_code_dict.items():
            if row['SpatialDim'] == value:
                country.append(key)
                break
    else:
        country.append("NF")
```

```python
ddata['Country'] = country
```

```python
ddata = ddata[['SpatialDim', 'TimeDim', 'NumericValue', 'Country']]
```

```python
ddata.to_csv('cleaned_data.csv')
```
